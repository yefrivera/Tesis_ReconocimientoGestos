```python
# Contenido para un cuaderno de Jupyter

# CELDA 1: Importaciones
import json
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# CELDA 2: Cargar un archivo de ejemplo
DATA_PATH = "../data/processed/mano_abierta"
sample_file = os.path.join(DATA_PATH, os.listdir(DATA_PATH)[5]) # Cargar el primer archivo

with open(sample_file, 'r') as f:
    gesture_data = json.load(f)

print(f"Datos del gesto: {gesture_data['gesture_name']}")
print(f"Número de frames: {len(gesture_data['frames'])}")
```

```python
# CELDA 3: Visualizar un frame
def plot_hand(frame_data):
    if not frame_data['landmarks']:
        print("No hay landmarks en este frame.")
        return
    
    landmarks = frame_data['landmarks']
    x = [lm['x'] for lm in landmarks]
    y = [-lm['y'] for lm in landmarks] # Invertir Y para visualización correcta

    plt.figure(figsize=(5, 5))
    plt.scatter(x, y)
    plt.title(f"Visualización del Frame {frame_data['frame']}")
    plt.xlabel("Coordenada X")
    plt.ylabel("Coordenada Y")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# Visualizar el primer frame con datos
for frame in gesture_data['frames']:
    if frame['landmarks']:
        plot_hand(frame)
        break
```
